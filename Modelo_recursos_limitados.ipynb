## Modelo con recursos limitados 

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
f0 = 1.9
s0 = 0.1
dt = 1/1000
T = 20
N = int(T/dt)

solucion_fish = np.zeros(N+1)
solucion_shark = np.zeros(N+1)

solucion_fish[0] = f0
solucion_shark[0] = s0

for n in range(N):
    solucion_fish[n+1] = solucion_fish[n] + dt*(2-solucion_fish[n]-solucion_shark[n])*solucion_fish[n]
    solucion_shark[n+1] = solucion_shark[n] + dt*(solucion_fish[n] - 1)*solucion_shark[n]

t = np.linspace(0, T, N+1)
plt.figure(figsize=(12, 8))
sns.lineplot(x=t, y=solucion_fish, color="blue", label="Peces")
sns.lineplot(x=t, y=solucion_shark, color="red", label="Tiburones")
plt.title("Modelo Lotka-Volterra con recursos limitados", fontweight="bold", fontsize=20)
plt.ylabel("Poblacion")
plt.xlabel("t")
plt.legend()
plt.show()
```

# Conclusiones 

Cuando $\beta=2$, significa que la capacidad de carga del entorno est limitada, es decir, que los recursos no permiten que la poblacion de peces crezca indefinidamente, incluso si no existen depredadores. Cuando se integra este elemento limitante al modelo, deja de observarse un comportamiento cclico en el tiempo, y en su lugar se puede observar un estado estacionario en el cual las poblaciones de ambas especies se mantienen constantes.
